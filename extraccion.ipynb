```python
#Filtro de filas vs columnas
#Paso 1: Leemos el contenido del archivo
from google.colab import files
uploaded=files.upload()
df=pd.read_csv(io.BytesIO(uploaded['CountryTable.csv']), index_col=11)
df
#Paso 2: Aplicamos filtro de filas vs columnas
filtro42=df.loc[["Republic","Socialistic Republic"], ["population"]]
filtro42 
#Aplicamos el filtro de filas para obtener republicas y republicas socialistas y en cuanto a columnas la poblaci√≥n
```

```python
#Filtro por cabecera
filtro51=df.head(4)
filtro51
#Filtramos los primeros cuatro datos
```

```python
#Filtro por cabecera
filtro52=df.head(3)
filtro52
#filtramos los primeros tres datos de la tabla
```

```python
#Filtro por comparacion
filtro61=df[df["continent"] == "Africa"]
filtro61
#filtramos la columna de continente para solo obtener la de Africa 
```

```python
#Filtro por comparacion
filtro62=df[df["name"] == "Angola"]
filtro62
#filtramos la columna name para solo obtener datos de Angola
```

```python
#Filtro Y
filtro71=df[(df["independence_year"] == 1960) & (df["continent"] =="Africa") & (df["region"] == "Central Africa")]
filtro71
#filtramos que el a√±o de independencia sea solo para 1960 que el continente sea solo Africa y que la region solo sea para Central Africa
```

```python
#Filtro Y
filtro72=df[(df["capital"] > 902) & (df["life_expectancy"]>50) & (df["gnp"] < 6000)]
filtro72
#Filtramos que el capital nos de solo para mayores a 902 al igual que life expentancy >50 y que GNP sea para menores de 6000
```

```python
#Filtro 8
filtro81=df[(df["continent"]== "Oceania") | (df["capital"] > 2000)]
filtro81
#Se filtraron todos los datos que fueran de oceania o que su capital fuera mayor a 2000
```

```python
#Filtro 8
filtro82=df[(df["life_expectancy"] <50) | (df["continent"]=="Oceania")]
filtro82
#Se obtuvieron todos los datos que tuvieran una expectativa de vida menor a 50 o que fueran de ocean√≠a
```

```python
#Filtro not
filtro91=df[~(df["continent"]=="Africa") & ~(df["head_of_state"]== "Elisabeth II")]
filtro91
#Todos los datos que no sean del continente de √Åfrica  y que la cabeza de estado no fuera Elisabeth II
```

```python
#Filtro not
filtro92=df[~(df["code"]=="COK") & ~(df["name"]== "Yugoslavia")]
filtro92
#Todos los datos que no fueran de c√≥digo COK y que tampoco fuera Yugoslavia
```
